```python
import pandas as pd
import numpy as np

train = pd.read_csv("C:/Users/Marielle/Documents/Uni Tartu/Introduction to Computational Neuroscience/traindata-20180502T082611Z-001/traindata/A01TInstances.csv", header=None)
labels = pd.read_csv("/Users/Marielle/Documents/Uni Tartu/Introduction to Computational Neuroscience/traindata-20180502T082611Z-001//traindata/A01TLabels.csv", header=None)

print(train.shape)
print(labels.shape)
```

```python
label_results = [0] * 672528

for index, row in labels.iterrows():
    label = row[0]
    row_nr = row[1]
    duration = row[2]
    for i in range(duration):
        current_label = label_results[row_nr + i - 1]
        if label == 769 or label == 770:
            if current_label == 769 and label == 770:  #Maybe there are cases where the label is 769 and 770 at the same time
                print("big problem1")
            elif current_label == 770 and label == 769:
                print("big problem2")
            label_results[row_nr + i - 1] = label
        elif current_label != 769 and current_label != 770:
            label_results[row_nr + i - 1] = 1

    
se = pd.Series(label_results)
train['label'] = se.values
```

```python
dd = train[train.label==769]
ss = train[train.label==770]
#ee = train[train.label==1]
result = dd.append(ss)
#result = result.append(ee)
```

```python
result
```

```python
result = result.dropna()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(result.iloc[:,0:25], result.iloc[:,-1], test_size=0.2, random_state=0)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda_clf = LDA()
lda_clf.fit(X_train, y_train)
```

```python
y_pred = lda_clf.predict(X_test)
count = np.sum(y_pred == y_test.values)
acc = count / len(y_pred)
acc
```

```python
from sklearn.metrics import precision_score
df = {'Precisision': precision_score(y_test.values, y_pred, average=None)}
df = pd.DataFrame(data=df)
df
```

```python
from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix(y_test, y_pred, labels=[770,769])
y_pred
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train) 
```

```python
y_pred_knn = model.predict(X_test)
count = np.sum(y_pred_knn == y_test.values)
acc = count / len(y_pred_knn)
acc
```
