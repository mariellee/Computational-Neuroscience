```python
import pandas as pd
import numpy as np

train = pd.read_csv("C:/Users/Marielle/Documents/Uni Tartu/Introduction to Computational Neuroscience/traindata-20180502T082611Z-001/traindata/A01TInstances.csv", header=None)
labels = pd.read_csv("/Users/Marielle/Documents/Uni Tartu/Introduction to Computational Neuroscience/traindata-20180502T082611Z-001//traindata/A01TLabels.csv", header=None)

print(train.shape)
print(labels.shape)
```

```python
label_results = [0] * 672528

for index, row in labels.iterrows():
    label = row[0]
    row_nr = row[1]
    #duration = row[2]
    if(row[0] == 770 or row[0] == 769 or row[0] == 771 or row[0] == 772):
        duration = 1000
    for i in range(duration):
        current_label = label_results[row_nr + i - 1]
        if label == 769 or label == 770 or label == 771 or label == 772:
            
            label_results[row_nr + i - 1] = label
        elif current_label != 769 and current_label != 770 and current_label != 771 and current_label != 772:
            label_results[row_nr + i - 1] = 1

    
se = pd.Series(label_results)
train['label'] = se.values
```

```python
class1 = train[train.label==769]
class2 = train[train.label==770]
class3 = train[train.label==771]
class4 = train[train.label==772]

result = class1.append(class2)
result = result.append(class3)
result = result.append(class4)
```

```python
result = result.reset_index()
df = result.drop(columns=['index'])
```

```python
last_index = 0
fft_array_rows = []

for i in range(288):
    current_matrix = df[last_index:last_index+1000]
    last_index = last_index + 1000
    fft_array_columns = []

    for i in range(25):
        fft_array_columns.append(np.fft.fft(current_matrix[i],1)[0])
        
    fft_array_rows.append(fft_array_columns)
        
    
```

```python
result_2 = pd.DataFrame(fft_array_rows)
```

```python
result_2
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(result_2.iloc[:,0:25], result_2.iloc[:,-1], test_size=0.2, random_state=0)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda_clf = LDA()
lda_clf.fit(X_train, y_train)
```

```python
y_pred = lda_clf.predict(X_test)
count = np.sum(y_pred == y_test.values)
acc = count / len(y_pred)
acc
```

```python
from sklearn.metrics import precision_score
df = {'Precisision': precision_score(y_test.values, y_pred, average=None)}
df = pd.DataFrame(data=df)
df
```

```python
from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix(y_test, y_pred, labels=[770,769])
y_pred
```
